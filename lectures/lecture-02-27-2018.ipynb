```matlab
A = [ 1 1 0 3
      2 1 -1 1
      3 -1 -1 2
      -1 2 3 -1 ];
  
B = [ 4 ; 1; -3; 4 ];
X = [];
 
% Remove x1 from P2, P3 and P4
dim = size(A);
m = dim(1);

for k = 1:m-1
    for i=k+1:m
        lambda = A(i,k) / A(k,k);
        for j = k:4
            A(i,j) = A(i,j) - lambda * A(k,j);
        end
        B(i) = B(i) - lambda * B(k);
    end
end

% Reverse substitution
X(m) = B(m) / A(m,m);

for i=m-1:-1:1
    sum = 0;
    for j = i+1:m
        sum = sum + (A(i,j) * X(j));
    end
    X(i) = (B(i) - sum) / A(i,i);
end

A
B
X
```

```matlab
A = [ 1 -1 2 -1
      2 -2 3 -3
      1 1 1 0
      1 -1 4 3 ];
  
B = [ 4 ; 1; -3; 4 ];
X = [];
 
% Remove x1 from P2, P3 and P4
dim = size(A);
m = dim(1);

for k = 1:m-1
    for i=k+1:m
        lambda = A(i,k) / A(k,k);
        for j = k:4
            A(i,j) = A(i,j) - lambda * A(k,j);
        end
        B(i) = B(i) - lambda * B(k);
    end
    % Check diagonal element, if it is 0, change with element beneath it
    s = 2;
    while A(k+1, k+1) == 0
        for p = 1:m
            temp = A(k+1, p);
            A(k+1, p) = A(k+s, p);
            A(k+s, p) = temp;
        end

        temp = B(k+1);
        B(k+1) = B(k+s);
        B(k+s) = temp;
        
        s = s + 1;
    end 
end

% Reverse substitution
X(m) = B(m) / A(m,m);

for i=m-1:-1:1
    sum = 0;
    for j = i+1:m
        sum = sum + (A(i,j) * X(j));
    end
    X(i) = (B(i) - sum) / A(i,i);
end

A
B
X
```
