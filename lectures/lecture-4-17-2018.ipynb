```matlab
% data
x = [ 1 ; 2 ; 5 ; 6 ; 7 ; 8 ; 9 ; 12 ; 13 ; 14 ; 15 ; 16 ; 19 ; 20 ; 21 ; 22 ; 23 ; 26 ; 27 ; 28 ; 29 ]; 
y = [ 13793 ; 13746 ; 13740 ; 13750 ; 13763 ; 13774 ; 13794 ; 13768 ; 13757 ; 13739 ; 13748 ; 13765 ; 13765 ; 13761 ; 13759 ; 13737 ; 13780 ; 13776 ; 13708 ; 13745 ; 13756];

% banyak data
dim = size(x);
N = dim(1);

% nilai x interpolasi
xmin = x(1);
xmax = x(N);
Nt = 1001;
dx = (xmax - xmin) / (Nt - 1);

for k = 1:Nt+1
    xi(k) = xmin + (k-1) * dx;
    % interpolasi
    sum = 0.0;
    for i = 1:N
        prod = 1.0;
        for j = 1:N
            if i~=j
                prod = prod * (xi(k) - x(j)) / (x(i) - x(j));
            end
        end
        L(i) = prod;
        sum = sum + L(i) * y(i);
    end
    P(k) = sum;
end

plot(x,y,'o', xi, P, 'r')

```
